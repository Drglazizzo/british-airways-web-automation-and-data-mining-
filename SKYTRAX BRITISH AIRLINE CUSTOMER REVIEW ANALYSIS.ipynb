```python
# Cell 1: Import necessary libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import os
import nltk
from nltk.corpus import stopwords
import re
from textblob import TextBlob
from gensim import corpora, models
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from pptx import Presentation

print("Libraries imported.")
```

```python
def open_driver(url):
    # Configure Chrome options
    options = webdriver.ChromeOptions()
    options.add_experimental_option('excludeSwitches', ['enable-logging'])  # Ignore warnings and errors
    
    # Initialize the WebDriver
    driver = webdriver.Chrome(options=options)
    
    # Navigate to the specified URL
    driver.get(url)
    
    # Set an implicit wait
    driver.implicitly_wait(20)  # Set a waiting time limit for finding elements
    driver.maximize_window()  # Maximize the browser window
    
    print("Driver opened and navigated to:", url)
    return driver
```

```python
# Cell 3: Scrape all pages of Skytrax reviews

base_url = "https://www.airlinequality.com/airline-reviews/british-airways/page/"

reviews_data = []
page_num = 1

while True:
    url = f"{base_url}{page_num}"
    print(f"Scraping page {page_num}...")

    driver = open_driver(url)
    time.sleep(random.uniform(2, 4))  # Random delay between requests

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = soup.find_all('article', {'itemtype': 'http://schema.org/Review'})

    if not articles:
        print(f"No more reviews found on page {page_num}. Ending scrape.")
        driver.quit()
        break

    for article in articles:
        try:
            review_text = article.find('div', class_='text_content').get_text(strip=True)
            rating = None
            rating_div = article.find('div', class_='rating-10')
            if rating_div:
                rating = rating_div.find_all('span')[1].text.strip()
            date = article.find('time')['datetime'] if article.find('time') else None

            reviews_data.append({
                'Review': review_text,
                'Rating': rating,
                'Date': date
            })
        except Exception as e:
            continue  # Skip malformed entries

    driver.quit()
    page_num += 1

# Save raw data
df_raw = pd.DataFrame(reviews_data)
os.makedirs("data", exist_ok=True)
df_raw.to_csv("data/raw_reviews.csv", index=False)
print("Raw reviews saved to data/raw_reviews.csv")
```

```python
# Cell 4: Clean the scraped text data

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower().split()
    return " ".join([word for word in text if word not in stop_words])

df = pd.read_csv("data/raw_reviews.csv")
df['Cleaned_Review'] = df['Review'].apply(clean_text)

df.to_csv("data/cleaned_reviews.csv", index=False)
print("Data cleaned and saved to data/cleaned_reviews.csv")
```

```python
# Cell 5: Analyze sentiment and extract topics

# Sentiment Analysis
df['Sentiment'] = df['Cleaned_Review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['Sentiment_Label'] = df['Sentiment'].apply(
    lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral')
)

# Topic Modeling (LDA)
texts = [row.split() for row in df['Cleaned_Review']]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)
topics = lda_model.print_topics(num_words=5)

# Save results
df.to_csv("data/analyzed_reviews.csv", index=False)

with open("data/topics.txt", "w") as f:
    for idx, topic in enumerate(topics):
        f.write(f"Topic {idx}: {topic}\n")

print("Analysis complete. Files saved:")
print("- analyzed_reviews.csv")
print("- topics.txt")
```

```python
# Cell 6: Create visualizations

# Sentiment Distribution Chart
sentiment_counts = df['Sentiment_Label'].value_counts()
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Customer Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.savefig('data/sentiment_chart.png')
plt.close()

# Word Cloud
all_text = ' '.join(df['Cleaned_Review'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('data/wordcloud.png')
plt.close()

print("Visualizations created:")
print("- sentiment_chart.png")
print("- wordcloud.png")
```

```python
# Cell 7: Create PowerPoint summary slide

prs = Presentation()
slide = prs.slides.add_slide(prs.slide_layouts[5])
title = slide.shapes.title
title.text = "British Airways Customer Review Insights"

# Add sentiment chart
slide.shapes.add_picture('data/sentiment_chart.png', 0, 100, width=400, height=300)

# Add word cloud
slide.shapes.add_picture('data/wordcloud.png', 400, 100, width=400, height=300)

# Add key insights
textbox = slide.shapes.add_textbox(0, 400, 800, 100)
tf = textbox.text_frame
tf.text = "Key Insights:"
p1 = tf.add_paragraph()
p1.text = "- Majority of reviews are positive, but recurring complaints about staff service and delays."
p2 = tf.add_paragraph()
p2.text = "- Common topics: Baggage Handling, Seat Comfort, Lounge Facilities, Flight Delays."

# Save presentation
prs.save('data/BA_Customer_Review_Summary.pptx')
print("PowerPoint summary created: data/BA_Customer_Review_Summary.pptx")
```

```python
from collections import Counter
from pptx import Presentation
from pptx.util import Inches
from pptx.enum.text import PP_ALIGN
import seaborn as sns

# Split reviews by sentiment
positive_words = " ".join(df[df['Sentiment_Label'] == 'Positive']['Cleaned_Review'])
negative_words = " ".join(df[df['Sentiment_Label'] == 'Negative']['Cleaned_Review'])

# Count top words
def get_top_n_words(text, n=10):
    words = text.split()
    return Counter(words).most_common(n)

top_pos = get_top_n_words(positive_words, 10)
top_neg = get_top_n_words(negative_words, 10)

pos_df = pd.DataFrame(top_pos, columns=["Word", "Frequency"])
neg_df = pd.DataFrame(top_neg, columns=["Word", "Frequency"])

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Frequency", y="Word", data=pos_df, color='green', label='Positive')
sns.barplot(x="Frequency", y="Word", data=neg_df, color='red', label='Negative')

plt.title('Top Words by Sentiment')
plt.legend()
plt.tight_layout()
plt.savefig('top_words_by_sentiment.png')
plt.close()
plt.show()
```
